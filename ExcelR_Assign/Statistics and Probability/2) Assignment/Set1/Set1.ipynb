### Outliers Detecting

```python
Outliers = []

def detect_outliers(data):
    q1 = data.quantile(.25)
    q3 = data.quantile(.75)
    iqr = q3 - q1
    
    upper_taill = q3 + 1.5 * iqr
    lower_tail = q1 - 1.5 * iqr
    
    for out in data:
        if out > upper_taill or out < lower_tail:
            Outliers.append(out)
            print(Outliers)
detect_outliers(df['Measure_X'])
```
