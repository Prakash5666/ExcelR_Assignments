# Q.11)

#### Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?

```python
conf_94 =stats.t.interval(alpha = 0.94, df=1999, loc=200, scale=30/np.sqrt(2000))
print('94% confidence interval :',conf_94)
conf_98= stats.t.interval(alpha = 0.98, df = 1999, loc = 200, scale=30/np.sqrt(2000))
print('98% confidence interval :',conf_98)
conf_96 = stats.t.interval( alpha = 0.96, df = 1999 , loc = 200 , scale = 30/np.sqrt(2000))
print('96% confidence interval :',conf_96)
```

# Q.20)

#### Calculate probability from the given dataset for the below cases

```python
Q20 = pd.read_csv('Cars.csv')
Q20.head()
```

```python
Q20.describe()
```

#### 1) P(MPG>38)

```python
MPG_is_gretter_38 = np.round(1 - stats.norm.cdf(38, loc= Q20.MPG.mean(), scale= Q20.MPG.std()),3)
print('P(MPG > 38) =',MPG_is_gretter_38)
```

#### 2) P(MPG<40)

```python
MPG_is_less_40 = np.round(stats.norm.cdf(40, loc = Q20.MPG.mean(), scale = Q20.MPG.std()),3)
print('P(MPG < 40)=',MPG_is_less_40)
```

#### 3) P (20<MPG<50)

```python
MPG_is_greater_20 = np.round(1 - stats.norm.cdf(20,loc = Q20.MPG.mean(),scale = Q20.MPG.std()),3)
MPG_is_less_50 = np.round(stats.norm.cdf(50, loc = Q20.MPG.mean(),scale = Q20.MPG.std()),3)
MPG_is_greater_20_and_less_50 = (MPG_is_less_50) - (MPG_is_greater_20)
print('P(20 < MPG < 50) =',(MPG_is_greater_20_and_less_50))
```
